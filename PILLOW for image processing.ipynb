# Copying

```python
copy=image.copy()
plt.imshow(copy)
```

```python
T1 = image.transpose(Image.FLIP_LEFT_RIGHT)
T2 = image.transpose(Image.FLIP_TOP_BOTTOM)
T3 = image.transpose(Image.ROTATE_90)
T4 = image.transpose(Image.ROTATE_180)
T5 = image.transpose(Image.TRANSPOSE)
plt.figure(figsize=(10,10))
plt.subplot(3,2,1)
plt.imshow(T1)
plt.title("FLIP_LEFT_RIGHT")
plt.subplot(3,2,2)
plt.imshow(T2)
plt.title("FLIP_TOP_BOTTOM")
plt.subplot(3,2,3)
plt.imshow(T3)
plt.title("ROTATE_90")
plt.subplot(3,2,4)
plt.imshow(T4)
plt.title("ROTATE_180")
plt.subplot(3,2,5)
plt.imshow(T1)
plt.title("TRANSPOSE")
```

```python
newsize=(300,300)
plt.figure(figsize=(10,10))
R1 = image.resize(newsize,Image.BILINEAR)
R2 = image.resize(newsize,Image.NEAREST)
R3 = image.resize(newsize,Image.BOX)
R4 = image.resize(newsize,Image.HAMMING)
R5 = image.resize(newsize,Image.BICUBIC)
R6 = image.resize(newsize,Image.LANCZOS)
plt.subplot(3,2,1)
plt.imshow(R1)
plt.title("BILINEAR")
plt.subplot(3,2,2)
plt.imshow(R2)
plt.title("NEAREST")
plt.subplot(3,2,3)
plt.imshow(R3)
plt.title("BOX")
plt.subplot(3,2,4)
plt.imshow(R4)
plt.title("HAMMING")
plt.subplot(3,2,5)
plt.imshow(R5)
plt.title("BICUBIC")
plt.subplot(3,2,6)
plt.imshow(R6)
plt.title("LANCZOS")
          
```

# Image Enchancement 

```python
from PIL import ImageEnhance
plt.figure(figsize=(10,10))
image_color = image.copy()
im1 = ImageEnhance.Color(image_color).enhance(2.5)
im2 = ImageEnhance.Contrast(image_color).enhance(2.5)
im3 = ImageEnhance.Brightness(image_color).enhance(2.5)
im4 = ImageEnhance.Sharpness(image_color).enhance(2.5)
plt.subplot(2,2,1)
plt.imshow(im1)
plt.title("color")
plt.subplot(2,2,2)
plt.imshow(im2)
plt.title("contrast")
plt.subplot(2,2,3)
plt.imshow(im3)
plt.title("brightness")
plt.subplot(2,2,4)
plt.imshow(im4)
plt.title("sharpness")
```

# Alpha blending

```python
image3 = image.copy()
image2 = image1.copy()

#Making both image same size
image2 = image2.resize(image3.size)

#blending 
blend = Image.blend(image3,image2,0.7)
plt.imshow(blend)
```
